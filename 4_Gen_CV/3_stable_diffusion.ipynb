```python
from diffusers import StableDiffusionPipeline
from diffusers import StableDiffusionImg2ImgPipeline
import torch
from PIL import Image
import requests
from io import BytesIO
from diffusers import AutoPipelineForInpainting
from diffusers.utils import load_image, make_image_grid
```

```python
#Text-to-Image

pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",  # یا CompVis/stable-diffusion-v1-4
    torch_dtype=torch.float16
)
pipe = pipe.to("cuda")
prompt = "a fantasy landscape with floating islands, vibrant colors, concept art"
image = pipe(prompt).images[0]
image.save("fantasy_landscape.png")
```

```python
prompt2 = "A serene lakeside at sunset, reflecting the vibrant colors of the sky, devoid of any human presence"
image2 = pipe(prompt).images[0]
image2.save("img2.png")
```

```python
#Image-to-Image

pipe = StableDiffusionImg2ImgPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5", torch_dtype=torch.float16
).to("cuda")

url = "https://raw.githubusercontent.com/CompVis/stable-diffusion/main/assets/stable-samples/img2img/sketch-mountains-input.jpg"
init_image = Image.open(BytesIO(requests.get(url).content)).convert("RGB")
init_image = init_image.resize((512, 512))
prompt = "a beautiful fantasy landscape, concept art, epic view"
result = pipe(prompt=prompt, image=init_image, strength=0.7, guidance_scale=7.5).images[0]
#result.show()
result.save("fantasy_result.png")
```

```python
# Inpainting
from diffusers import StableDiffusionInpaintPipeline
import torch
from PIL import Image
import requests
from io import BytesIO
pipe = StableDiffusionInpaintPipeline.from_pretrained(
    "stabilityai/stable-diffusion-2-inpainting", torch_dtype=torch.float16
).to("cuda")

image_url = "https://huggingface.co/datasets/huggingface/documentation-images/resolve/main/diffusers/inpaint.png"
mask_url = "https://huggingface.co/datasets/huggingface/documentation-images/resolve/main/diffusers/inpaint_mask.png"

init_image = Image.open(BytesIO(requests.get(image_url).content)).convert("RGB").resize((512, 512))
mask_image = Image.open(BytesIO(requests.get(mask_url).content)).convert("RGB").resize((512, 512))


prompt = "a cute tiger face, digital art"
image = pipe(prompt=prompt, image=init_image, mask_image=mask_image).images[0]

#image.show()
image.save("inpainted_result.png")

```

```python
#Style Transfer
pipe = StableDiffusionImg2ImgPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5", torch_dtype=torch.float16
).to("cuda")

url = "https://raw.githubusercontent.com/CompVis/stable-diffusion/main/assets/stable-samples/img2img/sketch-mountains-input.jpg"
image = Image.open(BytesIO(requests.get(url).content)).convert("RGB").resize((512, 512))
prompt = "a landscape painting in the style of Vincent van Gogh"
result = pipe(prompt=prompt, image=image, strength=0.8, guidance_scale=8.0).images[0]
result.show()
result.save("st_result.png")
```
